```python
import pandas as pd

# Load the race-category.csv file
file_path = 'race category.csv'
race_data = pd.read_csv(file_path, header=None, names=['Category'])
race_data.head()
```

```python
import re

# Drop the first row which is the header and any NaN values
race_data_clean = race_data.dropna().iloc[1:].reset_index(drop=True)

# Define age groups
age_groups = [(18,24), (25,29), (30,39), (40,49), (50,59), (60,69), (70,79), (80,89), (90,99)]

# Function to map category to age group

def map_to_age_group(category):
    # Extract all numbers from the category string
    numbers = list(map(int, re.findall(r'\d+', category)))
    if not numbers:
        return 'Other'
    
    # Check if any number falls into the defined age groups
    for start, end in age_groups:
        for num in numbers:
            if start <= num <= end:
                return f'{start}-{end}'
    return 'Other'

# Apply the mapping function
race_data_clean['Age Group'] = race_data_clean['Category'].apply(map_to_age_group)
race_data_clean.head(20)
```
